```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from IPython.display import display
%matplotlib inline
```

```python
#wineのデータセットを読み込む(sklearnに含まれている)
from sklearn.datasets import load_wine
wine_dataset = load_wine()
```

```python
print(wine_dataset["DESCR"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    wine_dataset['data'], wine_dataset['target'], random_state=0)
```

```python
wine_dataframe = pd.DataFrame(X_train, columns=wine_dataset.feature_names)
```

```python
display(wine_dataframe)
```

```python
grr = pd.plotting.scatter_matrix(wine_dataframe, c=y_train, figsize=(15,15), marker=".",
                                hist_kwds={"bins": 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```
